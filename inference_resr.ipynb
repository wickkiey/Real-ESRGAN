```python
model = RRDBNet(num_in_ch=3, num_out_ch=3, num_feat=64, num_block=23, num_grow_ch=32, scale=4)
netscale = 4
# file_url = ['https://github.com/xinntao/Real-ESRGAN/releases/download/v0.1.0/RealESRGAN_x4plus.pth']
```

```python
# model_path  = r"./weights/RealESRGAN_x4plus.pth"
model_path = r"D:\wickkiey_git\Real-ESRGAN\experiments\debug_finetune_RealESRGANx4plus_vb\models\net_g_latest.pth"
```

```python
upsampler = RealESRGANer(
        scale=netscale,
        model_path=model_path,
        dni_weight=None,
        model=model,
        tile=0,
        tile_pad=10,
        pre_pad=0,
        half=not True,
        gpu_id=None)
```

```python
face_enhancer = GFPGANer(
            model_path='https://github.com/TencentARC/GFPGAN/releases/download/v1.3.0/GFPGANv1.3.pth',
            upscale=4,
            arch='clean',
            channel_multiplier=2,
            bg_upsampler=upsampler)
```

```python
# read video frames and pass to face enhancer and save back to video

input_file = "./inputs/video/Veerabagu.mp4"
output_file = "./inputs/video/vb_images/"
```

```python
# read opencv video and write back with best compression quality

cap = cv2.VideoCapture(input_file)
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
counter = 0
while cap.isOpened():
    ret, frame = cap.read()
    if ret:
        frame = cv2.resize(frame, (width//4, height//4))
        
        _, _, output = face_enhancer.enhance(frame, has_aligned=False, only_center_face=False, paste_back=True)
        # output = cv2.resize(frame, (width, height))
        print(output.shape)
        cv2.imwrite(output_file + str(counter) + ".jpg", output)
        counter += 1
        # break
    else:
        break
```
